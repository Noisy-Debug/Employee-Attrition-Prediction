```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
file_path = "1-DATA/HR-Employee-Attrition.csv"
df = pd.read_csv(file_path)
```

```python
# 1. Attrition Count
plt.figure(figsize=(8, 5), clear=True)
attrition_counts = df["Attrition"].value_counts()
ax = attrition_counts.plot(kind="bar", color=["skyblue", "salmon"], edgecolor='black')
for i, v in enumerate(attrition_counts):
    plt.text(i, v + 5, str(v), ha='center', fontsize=12, fontweight='bold')
plt.title("Employee Attrition Count", fontsize=14)
plt.xlabel("Attrition", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 2. Attrition vs. Monthly Income
plt.figure(figsize=(10, 6), clear=True)
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df, hue="Attrition", palette=["skyblue", "salmon"], legend=False)
plt.title("Attrition vs. Monthly Income", fontsize=14)
plt.xlabel("Attrition", fontsize=12)
plt.ylabel("Monthly Income", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Attrition by Department
plt.figure(figsize=(8, 5), clear=True)
df.groupby("Department")["Attrition"].value_counts().unstack().plot(kind="bar", stacked=True, color=["lightgreen", "tomato"], edgecolor='black')
plt.title("Attrition by Department", fontsize=14)
plt.xlabel("Department", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Attrition", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 4. Attrition by Job Role
plt.figure(figsize=(12, 6), clear=True)
df.groupby("JobRole")["Attrition"].value_counts().unstack().plot(kind="bar", stacked=True, color=["cornflowerblue", "indianred"], edgecolor='black')
plt.title("Attrition by Job Role", fontsize=14)
plt.xlabel("Job Role", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Attrition", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 5. Work-Life Balance vs Attrition
plt.figure(figsize=(8, 6), clear=True)
ax = sns.countplot(data=df, x="WorkLifeBalance", hue="Attrition", palette=["skyblue", "salmon"], edgecolor='black')
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width()/2., height + 2, int(height), ha="center", fontsize=12, fontweight="bold")

plt.title("Work-Life Balance vs Attrition", fontsize=14)
plt.xlabel("Work-Life Balance (1 = Low, 4 = High)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Attrition")
plt.show()
```

```python
# 6. Job Satisfaction vs. Attrition
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x="JobSatisfaction", hue="Attrition", palette=["lightblue", "tomato"], edgecolor='black')
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width()/2., height + 2, int(height), ha="center", fontsize=12, fontweight="bold")

plt.title("Job Satisfaction vs. Attrition", fontsize=14)
plt.xlabel("Job Satisfaction (1 = Low, 4 = High)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Attrition")
plt.show()
```
